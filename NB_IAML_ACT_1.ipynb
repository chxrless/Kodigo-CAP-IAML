**Observaciones:**

**Limpiando Columnas Irrelevantes...**

```python
df.loc[:,['DR_NO','Date Rptd','DATE OCC','TIME OCC','AREA','AREA NAME','Crm Cd','Crm Cd Desc','Vict Age','Vict Sex','Vict Descent','Weapon Used Cd','Status','Crm Cd 1','LOCATION']]
```

**Formato de Fechas...**

```python
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'], format = '%m/%d/%Y %I:%M:%S %p')
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'], format = '%m/%d/%Y %I:%M:%S %p')
```

**Quitando espacios en LOCATION...**

```python
df['LOCATION'] = df['LOCATION'].str.strip()
```

### Análisis

Vamos a ver el comportamiento de crímenes a lo largo de los meses transcurridos del 2024,  
para eso vamos a hacer un dataframe agrupado por mes

```python
#Columna nueva para obtener solamente el mes del suceso
df['month_occ'] = df['DATE OCC'].dt.month

#Agupamos...
df_mensual = df.groupby(['month_occ'])
```

Ahora veamos gráficamente el índice de crímenes por mes durante el 2024

```python
import seaborn as sns
sns.set_theme()
sns.set(rc = {"figure.dpi" : 300})
import warnings
warnings.filterwarnings("ignore")
```

```python
df_mensual.size().plot(kind = 'bar')
```

... Interesante, podemos observar que los crímenes han tendido a bajar respecto van avanzando los meses. Una buena noticia :)  
Visto en una tabla, tenemos los siguientes datos:

```python
print(df_mensual['month_occ'].count())
```

Ahora vamos a ver cuáles son los tipos de crímenes más cometidos...  
Debido a que son muchas las categorías, vamos a limitarlo al top 5 de todo el universo

```python
df_crimenes = df.groupby(['Crm Cd Desc']).size().reset_index(name = 'count')
df_crimenes_t5 = df_crimenes.nlargest(5,'count')
```

```python
sns.barplot(x = 'Crm Cd Desc', y = 'count', data = df_crimenes_t5)
```

... Visto en tabla:

```python
print(df_crimenes_t5)
```

Podemos notar que el crimen más recurrente es el robo de vehículos, con 10990 veces durante todo lo recorrido en el 2024.  
Las demás tipologías se quedan por debajo de la mitad.

Ahora veremos cuál es el grupo social más afectado dentro de los crímenes cometidos

```python
df_g_social = df.groupby(['Vict Sex'])
```

```python
df_g_social.size().plot(kind = 'bar')
```

```python
print(df_g_social.size())
```

Podemos observar que los más afectados son las personas de sexo masculino

## Conclusión  
Con los análisis realizados, podemos concluir que los índices de crímenes en Los Ángeles, California, tienden a la baja en lo que va del año 2024.  
Además, el crimen más frecuente son los robos de vehículos y las personas más afectadas son los de sexo masculino.
